# Confusion Matrix

```python
from sklearn import metrics
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
sz = 100
nclass = 4
y = np.random.randint(nclass, size=sz)
yhat = np.random.randint(nclass, size=sz)
classes = list(range(nclass))
```

```python
cm = metrics.confusion_matrix(y, yhat)
cmp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
fig, ax = plt.subplots(figsize=(10,10))
cmp.plot(ax=ax)
```

```python
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
diag = cm_norm.diagonal() # get diagonal
```

```python
for i, p in enumerate(diag):
    print(i, classes[i], p)
```
